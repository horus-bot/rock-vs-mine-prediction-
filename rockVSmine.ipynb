```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonarData = pd.read_csv('/content/sonar.all-data.csv' , header=None)
```

```python
sonarData.head()
```

```python
sonarData.describe()
sonarData.shape
```

```python
sonarData[60].value_counts()
```

```python
X = sonarData.drop( columns= 60 , axis=1)
Y = sonarData[60]
print(X)
print(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split( X, Y , test_size=0.1 , stratify=Y , random_state=1)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
accuracy = accuracy_score(X_train_prediction , Y_train)
print(accuracy)
```

```python
X_test_prediction = model.predict(X_test)
accuracyTest = accuracy_score(X_test_prediction , Y_test)
print(accuracyTest)
```
